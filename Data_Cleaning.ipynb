```python
#정확성(이상점-제거하거나or유의미하면 놔두기)
df = pd.read_csv('beer.csv', index_col=0)
df.head()
```

```python
df.plot(kind='box', y='abv')
```

```python
df['abv'].describe()
```

```python
df['abv'].quantile(0.25) #25%지점
```

```python
df['abv'].quantile(0.75) #75%지점
```

```python
q1 = df['abv'].quantile(0.25) #25%지점
q3 = df['abv'].quantile(0.75) #75%지점
iqr = q3 - q1
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 *iqr ) #이상점
```

```python
df[condition]
```

```python
df.loc[2250, 'abv'] = 0.055
```

```python
df.loc[2250]
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 *iqr ) #이상점
df[condition]
```

```python
df[condition].index
```

```python
df.drop(df[condition].index, inplace=True)
```

```python
condition = (df['abv'] < q1 - 1.5 * iqr) | (df['abv'] > q3 + 1.5 *iqr ) #이상점
df[condition]
```
